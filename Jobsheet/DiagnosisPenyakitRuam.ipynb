# **DIAGNOSIS PENYAKIT BERDASARKAN GEJALA RUAM MENGGUNAKAN BAYES THEOREM**

## **Pendahuluan**

Dalam sistem berbasis pengetahuan, khususnya pada domain medis, proses diagnosis sering melibatkan ketidakpastian. Oleh karena itu digunakan pendekatan probabilistik berbasis *Bayes theorem* untuk memperbarui keyakinan terhadap suatu hipotesis penyakit berdasarkan evidence atau gejala yang diamati.

Teorema Bayes dinyatakan sebagai:

**$$P(H|E) = (P(E|H) * P(H)) / P(E)$$**

di mana:
- H adalah hipotesis, misalnya penyakit flu
- E adalah evidence, misalnya gejala ruam

Pendekatan ini memungkinkan sistem untuk menghitung peluang posterior, yaitu peluang suatu penyakit setelah diketahui adanya gejala tertentu. Hal ini sangat penting karena gejala yang sama dapat muncul pada beberapa penyakit dengan probabilitas yang berbeda.

Pada kasus ini, diketahui distribusi penyakit pada anak-anak serta peluang munculnya ruam pada masing-masing penyakit. Tujuannya adalah menghitung kemungkinan seorang anak menderita flu jika diketahui mengalami ruam.

### **1. Inisialisasi Data Probabilitas**

```python
# Prior probability
P_flu = 0.90
P_campak = 0.10

# Likelihood
P_ruam_given_flu = 0.08
P_ruam_given_campak = 0.95

print("=== DATA PROBABILITAS ===")
print(f"P(Flu) = {P_flu}")
print(f"P(Campak) = {P_campak}")
print(f"P(Ruam|Flu) = {P_ruam_given_flu}")
print(f"P(Ruam|Campak) = {P_ruam_given_campak}")
```

### **2.Visualisasi Formula Bayes**

```python
print("=== FORMULA BAYES ===")
print("P(Flu|Ruam) = [P(Ruam|Flu) * P(Flu)] / P(Ruam)")
```

### **3. Perhitungan P(Ruam) (Total Probability)**

```python
P_ruam = (P_ruam_given_flu * P_flu) + (P_ruam_given_campak * P_campak)

print("=== PERHITUNGAN P(Ruam) ===")
print(f"P(Ruam) = ({P_ruam_given_flu} * {P_flu}) + ({P_ruam_given_campak} * {P_campak})")
print(f"P(Ruam) = {P_ruam:.4f}")
```

### **4. Hitung P(Flu|Ruam)**

```python
P_flu_given_ruam = (P_ruam_given_flu * P_flu) / P_ruam

print("=== PERHITUNGAN P(Flu|Ruam) ===")
print(f"P(Flu|Ruam) = ({P_ruam_given_flu} * {P_flu}) / {P_ruam}")
print(f"P(Flu|Ruam) = {P_flu_given_ruam:.4f}")
```

### **5. Interpretasi Hasil**

```python
print("=== INTERPRETASI ===")
print(f"Jika seorang anak mengalami ruam, maka peluang anak tersebut terkena flu adalah:")
print(f"{P_flu_given_ruam:.4f} atau {P_flu_given_ruam*100:.2f}%")
```

### **6. Perbandingan dengan Campak**

```python
P_campak_given_ruam = (P_ruam_given_campak * P_campak) / P_ruam

print("=== PERBANDINGAN POSTERIOR ===")
print(f"P(Flu|Ruam) = {P_flu_given_ruam:.4f}")
print(f"P(Campak|Ruam) = {P_campak_given_ruam:.4f}")

if P_campak_given_ruam > P_flu_given_ruam:
    print("\nKesimpulan: Campak lebih mungkin dibanding flu jika terdapat ruam")
else:
    print("\nKesimpulan: Flu lebih mungkin dibanding campak")
```

### **7. Insight**

```python
print("=== INSIGHT ANALISIS ===")
print("Meskipun flu memiliki prior probability lebih tinggi,")
print("ruam adalah gejala yang jauh lebih kuat pada campak.")
print("Sehingga posterior probability campak menjadi lebih tinggi.")
```
