# **ANALISIS PELUANG KONDISIONAL PADA KETEPATAN WAKTU PENERBANGAN LION AIR**

## **Pendahuluan**

Dalam teori probabilitas, hubungan antar kejadian dapat dianalisis menggunakan konsep peluang kondisional (*conditional probability*), yang merupakan dasar dari *Bayes theorem*. Konsep ini digunakan untuk menghitung peluang suatu kejadian dengan mempertimbangkan informasi atau kondisi tertentu yang telah diketahui sebelumnya.

Secara matematis, peluang kondisional dirumuskan sebagai:

**$$P(A|B) = P(A ∩ B) / P(B)$$**

Dalam konteks sistem berbasis pengetahuan, pendekatan ini digunakan untuk melakukan *reasoning under uncertainty*, yaitu memperbarui keyakinan terhadap suatu kejadian berdasarkan informasi tambahan yang tersedia.

Studi kasus ini membahas ketepatan waktu penerbangan Lion Air dengan tiga parameter utama, yaitu peluang keberangkatan tepat waktu, peluang kedatangan tepat waktu, serta peluang keduanya terjadi secara bersamaan. Berdasarkan data tersebut, akan dihitung peluang kondisional untuk memahami hubungan dependensi antara keberangkatan dan kedatangan.

Pendekatan ini penting dalam analisis sistem transportasi dan juga menjadi dasar dalam pengembangan sistem pakar berbasis probabilistik.

### **1. Inisialisasi Data Probabilitas**

```python
# Probabilitas dasar
P_B = 0.85   # Berangkat tepat waktu
P_D = 0.90   # Datang tepat waktu
P_B_and_D = 0.75  # Berangkat dan datang tepat waktu

print("=== DATA PROBABILITAS ===")
print(f"P(B)  = {P_B}")
print(f"P(D)  = {P_D}")
print(f"P(B ∩ D) = {P_B_and_D}")
```

### **2.Visualisasi Konsep Probabilitas**

```python
print("=== KONSEP YANG DIGUNAKAN ===")
print("P(D|B) = P(B ∩ D) / P(B)")
print("P(B|D) = P(B ∩ D) / P(D)")
```

### **3. Perhitungan P(D|B)**

```python
P_D_given_B = P_B_and_D / P_B

print("=== PERHITUNGAN P(D|B) ===")
print(f"P(D|B) = {P_B_and_D} / {P_B}")
print(f"P(D|B) = {P_D_given_B:.4f}")
```

### **4. Interpretasi P(D|B)**

```python
print("=== INTERPRETASI ===")
print("Jika pesawat berangkat tepat waktu, maka peluang pesawat datang tepat waktu adalah:")
print(f"{P_D_given_B:.4f} atau {P_D_given_B*100:.2f}%")
```

### **5. Perhitungan P(B|D)**

```python
P_B_given_D = P_B_and_D / P_D

print("=== PERHITUNGAN P(B|D) ===")
print(f"P(B|D) = {P_B_and_D} / {P_D}")
print(f"P(B|D) = {P_B_given_D:.4f}")
```

### **6. Interpretasi P(B|D)**

```python
print("=== INTERPRETASI ===")
print("Jika pesawat datang tepat waktu, maka peluang pesawat berangkat tepat waktu adalah:")
print(f"{P_B_given_D:.4f} atau {P_B_given_D*100:.2f}%")
```

### **7. Perbandingan Hasil**

```python
print("=== PERBANDINGAN HASIL ===")
print(f"P(D|B) = {P_D_given_B:.4f}")
print(f"P(B|D) = {P_B_given_D:.4f}")

if P_D_given_B > P_B_given_D:
    print("\nKesimpulan: Ketepatan keberangkatan lebih kuat mempengaruhi ketepatan kedatangan")
else:
    print("\nKesimpulan: Ketepatan kedatangan lebih kuat mempengaruhi keberangkatan")
```

### **8. Insight Analisis**

```python
print("=== INSIGHT ANALISIS ===")
print("Nilai probabilitas menunjukkan adanya korelasi antara keberangkatan dan kedatangan.")
print("Namun, hubungan ini tidak simetris karena P(D|B) ≠ P(B|D).")
print("Hal ini menunjukkan bahwa keterlambatan masih dapat dikompensasi selama perjalanan.")
```

Pendekatan ini secara matematis merupakan bentuk sederhana dari Bayesian inference, di mana peluang diperbarui berdasarkan kondisi yang diketahui. Dalam sistem berbasis pengetahuan, metode ini digunakan untuk:

* diagnosis berbasis evidence
* prediksi berbasis probabilitas
* pengambilan keputusan di bawah ketidakpastian
