# **SIMULASI CERTAINTY FACTOR PADA DIAGNOSIS DEMAM BERDARAH**

Metode certainty factor digunakan untuk menghitung tingkat keyakinan terhadap suatu penyakit berdasarkan kombinasi gejala yang diberikan oleh user dan bobot dari pakar.

### **1. Input Data**

```python
# CF pakar
cf_pakar = [
    1.0, 1.0, 0.8, 0.6, 1.0, 0.6, 0.4, -0.4,
    0.4, 0.4, 1.0, 0.8, 0.6, -0.4, 1.0, 1.0
]

# CF user
cf_user = [
    0.4, 0.2, 0, 0.2, 0.8, 0.2, 0.6, 0.2,
    0, 0, 0.4, 0.4, 0, 0.4, 0.2, 0
]

print("=== DATA INPUT ===")
for i in range(len(cf_pakar)):
    print(f"Gejala {i+1}: CF_pakar={cf_pakar[i]}, CF_user={cf_user[i]}")
```

### **2. Hitung CF Partial**

```python
cf_values = []

print("=== CF PARSIAL ===")

for i in range(len(cf_pakar)):
    cf = cf_pakar[i] * cf_user[i]
    cf_values.append(cf)
    print(f"CF{i+1} = {cf:.4f}")
```

### **3. Fungsi Kombinasi CF**

```python
def combine_cf(cf1, cf2):
    return cf1 + cf2 * (1 - cf1)
```

### **4. Proses Kombinasi Iteratif**

```python
print("=== PROSES KOMBINASI ===")

cf_combine = cf_values[0]

for i in range(1, len(cf_values)):
    print(f"Step {i}: Combine {cf_combine:.4f} dengan {cf_values[i]:.4f}")
    cf_combine = combine_cf(cf_combine, cf_values[i])
    print(f"Hasil = {cf_combine:.4f}\n")
```

### **5. Hasil Akhir**

```python
print("=== HASIL AKHIR ===")
print(f"CF Demam Berdarah = {cf_combine:.4f}")
print(f"Persentase = {cf_combine * 100:.2f}%")
```

### **6. Interpretasi**

```python
print("=== INTERPRETASI ===")

if cf_combine > 0.8:
    print("Tingkat keyakinan sangat tinggi terhadap demam berdarah")
elif cf_combine > 0.6:
    print("Tingkat keyakinan cukup tinggi")
else:
    print("Tingkat keyakinan rendah")
```
